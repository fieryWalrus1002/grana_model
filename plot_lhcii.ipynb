```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
import numpy as np
import os
import matplotlib as mpl
from itertools import product
```

```python
combos = product([3.75, 4.5], range(4, 100, 20))
limit = 10000
```

```python
def get_random_color():
    return (np.random.random(), np.random.random(), np.random.random(), 1)

def get_patch(coords, x, y, theta):
    poly_coords = [(c[0] + x, c[1] + y) for c in coords]
    offset_coords = coords + np.array([x, y])
    t2 = mpl.transforms.Affine2D().rotate_around(x, y, theta) + ax.transData
    patch = Polygon(offset_coords, closed=True, transform=t2, fill=True, color=get_random_color())
    patch.set_transform(t2)
    return patch
    
def plot_patches(get_patch, coords, df, ax):
    for i, row in df.iterrows():
        x, y, theta = row['x'], row['y'], row['angle']
        ax.add_patch(get_patch(coords, x, y, theta))
```

```python


for combo in combos:
    radius, nodes = combo
    radius = str(radius)

    shape_file = f'src/grana_model/res/shapes/lhcii_circle_r_{radius}_n_{nodes}_coords.csv'
    lhcii_coords = f'C:/projects/grana_model/lhcii_export_coords/lhcii_circle_{radius.replace(".", "p")}_{nodes}_limit_{limit}_coords.csv'

    coords = pd.read_csv(shape_file).values
    df = pd.read_csv(lhcii_coords)

    fig, ax = plt.subplots(figsize=(10, 10))
    patches = []

    plot_patches(get_patch, coords, df, ax)

    plt.title(f'LHCII circles, radius = {radius}, nodes = {nodes}')

    ax.plot(df['x'], df['y'], color='forestgreen', linestyle='None')
    ax.set_aspect('equal')
    plt.savefig(f'shape_plots/lhcii_circle_r_{radius}_{nodes}_{limit}_colcheck.png')
```
